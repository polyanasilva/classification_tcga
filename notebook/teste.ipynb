```python
# Normalizacao log2(x+1)

df_log2 = np.log2(expressao_df + 1)
df_log2.head()

# tirar o log trab c dado bruto 
```

##  Filtragem por Baixa Express√£o ou Vari√¢ncia

```python
# FILTRO POR EXP MINIMA

limiar = 1        # valor m√≠nimo ap√≥s log2
frac_min = 0.2    # gene deve estar expresso em pelo menos 20% das amostras

# Criar m√°scara booleana ?? pesquisar oq isso
mask = (df_log2 > limiar).sum(axis=1) >= (frac_min * df_log2.shape[1])

# Filtrar genes
df_filtrado = df_log2[mask]

print(f"Genes antes do filtro: {df_log2.shape[0]}")
print(f"Genes depois do filtro: {df_filtrado.shape[0]}")
```

```python
df_filtrado.head()
```

```python
# Filtrar por vari√¢ncia

variancias = df_filtrado.var(axis=1)
medias = df_filtrado.mean(axis=1)

# Filtro baseado em limiares absolutos (dar uma olhada no Guo 2019)
mask = (variancias >= 1.0) & (medias >= 5.0)

print(f"Genes antes: {df_filtrado.shape[0]}")
df_filtr_novo = df_filtrado[mask]


print(f"Genes depois (vari√¢ncia>=1.0 e m√©dia>=5.0): {df_filtr_novo.shape[0]}")

# 3. Alternativa: remover os 25% com menor vari√¢ncia
percentil_25 = variancias.quantile(0.25)
mask_var = variancias > percentil_25
df_topvar = df_filtrado[mask_var]

print(f"Genes ap√≥s remover os 25% menos vari√°veis: {df_topvar.shape[0]}")
```
