```python
# Abrir o txt de genes mitocondriais
with open("../genes mito e ribo/genes_mitocondriais.txt", "r", encoding="utf-8") as f:
    linhas = f.readlines()

# Processar cada linha
linhas_processadas = []
for linha in linhas:
    # tira aspas, substitui espaço por vírgula e remove espaços extras
    linha_limpa = linha.replace('"', "").strip().replace(" ", ",")
    linhas_processadas.append(linha_limpa)

# Salvar em CSV
with open("../genes mito e ribo/genes_mitocondriais.csv", "w", encoding="utf-8") as f:
    for linha in linhas_processadas:
        f.write(linha + "\n")
```
